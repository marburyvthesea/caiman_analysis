```python
from PIL import Image
from skimage import io
import matplotlib.pyplot as plt
from tifffile import imsave
import glob
from tqdm import tqdm 
```

```python
for img_to_convert in tqdm(glob.glob('/projects/b1118/miniscope/data/GRIN044/1_26_2021/H13_M24_S21/*tif')):
    #load image as tiff stack
    tiff_stack = io.imread(img_to_convert)
    #split into r,g,b channels 
    r = tiff_stack[:, :, :, 0]
    g = tiff_stack[:, :, :, 1]
    b = tiff_stack[:, :, :, 2]
    #scale pixel values to gray
    gray_image = (r*0.299 + g*0.587 +b*0.114)
    imsave(img_to_convert, gray_image)
```

```python
plt.figure()
plt.imshow(gray_image[0,:,:])
```
